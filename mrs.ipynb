```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert_cast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies.head()
```

```python
movies.iloc[0].cast
```

```python
movies.iloc[0].crew
```

```python
def conver_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(conver_director)
```

```python
movies.iloc[0].crew
```

```python
movies.head()
```

```python
movies.iloc[0].overview
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
rq_movies = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
rq_movies.head()
```

```python
rq_movies['tags'] = rq_movies['tags'].apply(lambda x: " ".join(x))
```

```python
rq_movies.head()
```

```python
rq_movies.dtypes
```

```python
rq_movies['tags'] = rq_movies['tags'].str.lower()
```

```python
rq_movies.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stemm(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
 rq_movies['tags'].apply(stemm)
```

```python
rq_movies['tags'] =  rq_movies['tags'].apply(stemm)
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(rq_movies['tags']).toarray()
```

```python
get_feature_names= cv.get_feature_names_out()
```

```python
get_feature_names_df = pd.DataFrame(get_feature_names, columns=['Values'])
```

```python
get_feature_names_df
```

```python
get_feature_names_df.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
def recommend(movie):
    index = rq_movies[rq_movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(rq_movies.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
